```python
import pandas
from Corpus import Corpus
from Classes import Document
```

```python
df_discours =pandas.read_csv('discours_US.csv', sep ="\t")
df_discours.value_counts("speaker")
```

```python
corp = Corpus('TD8')
for index, row in df_discours.iterrows():
    phrases =row['text'].split(".")
    for k in phrases:
        if k != "":
            corp.add(Document(row["descr"],row['speaker'],row['date'],row['link'],k))
```

```python
corp.concorde('work')
```

```python
import SearchEngine as s
```

```python
engine = s.SearchEngine(corp)
```

```python
engine.search(100)
```

```python
import ipywidgets as widgets
from IPython.display import display
```

```python
button = widgets.Button(
    description='click to confirm the data',
    layout={'width': '300px'}
)
l =widgets.Label("Moteur de recherche")
t =widgets.Text()
i =widgets.IntSlider(50)
out = widgets.Output()
```

```python
def clique_bouton(event):
    query = t.value
    num = i.value
    with out:
        out.clear_output()
        # Call the search function
        results = engine.search(num, query)
        for result in results:
            print(result)


button.on_click(clique_bouton)
display(widgets.VBox([l,t,i,button,out]))
```
